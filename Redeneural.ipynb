#**Imports**

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

#**Data Base**

```python
#Carregar o arquivo
spam = pd.read_csv("/spam (1).csv")
spam.head()
```

```python
#transformar a coluna Category em variável numérica
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(spam['Category'])
print(y)
```

```python
#separar a coluna Message para treino e teste e a variável y em treino e teste
mensagens = spam['Message'].values
X_train, X_test, y_train, y_test = train_test_split(mensagens, y, test_size=0.3)
```

```python
X_train

#tokenizar as mensagens
#Processo de separar a sentença em suas partes: palavras, pontos, símbolos etc
token = Tokenizer(num_words=1000)
token.fit_on_texts(X_train)
token.fit_on_texts(X_train)
X_train = token.texts_to_sequences(X_train)
X_test = token.texts_to_sequences(X_test)


print(X_train)
```

```python
#Deixar as matrizes com o mesmo tamanho
X_train = pad_sequences(X_train, padding="post", maxlen=500)
X_test = pad_sequences(X_test, padding="post", maxlen=500)
```

```python
token.word_index
```

```python
#Nosso vocabulário de palavras
len(token.word_index)
```

```python
#Construir a rede neural
#Embedding é um vetor contextualizado dentro do texto
#Flatten liga a camada Embedding a camada Dense
#Dropout serve para reduzir o overfiting
#Funções Sigmóide e suas combinações geralmente funcionam melhor no caso de classificadores (saídas binárias).
#Funções Relu é utilizada nas camadas para redes profundas pois é computacionalmente mais simples (apenas um max(0, x)).

modelo = Sequential()
modelo.add(Embedding(input_dim=len(token.word_index),output_dim=50,input_length=500))
modelo.add(Flatten())
modelo.add(Dense(units=10, activation="relu"))
modelo.add(Dropout(0.1))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
```

```python
modelo.summary()
```

```python
modelo.fit(X_train, y_train, epochs=10, batch_size=10, verbose=True, validation_data=(X_test,y_test))
```

```python
loss, accuracy = modelo.evaluate(X_test,y_test)
print("Loss: ", loss)
print("Acurácia: ", accuracy)
```

```python
nova_previsao = modelo.predict(X_test)
print(nova_previsao)
```

```python
prev = (nova_previsao > 0.5)
print(prev)
```

```python
cm = confusion_matrix(y_test, prev)
print(cm)
```
